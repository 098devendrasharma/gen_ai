```python
import os
from dotenv import load_dotenv
load_dotenv()

os.environ["OPEN_API_KEY"] = os.getenv("OPEN_API_KEY")

os.environ["LANGCHAIN_API_KEY"] = os.getenv("LANGCHAIN_API_KEY") # this is use for langsmith tracking
os.environ["LANGCHAIN_TRACING_V2"] = "true"
os.environ["LANGCHAIN_PROJECT"] = os.getenv("LANGCHAIN_PROJECT") 
```

```python
from langchain_openai import ChatOpenAI
llm = ChatOpenAI()
print(llm)
```

```python
# How to call particular model 
from langchain_openai import ChatOpenAI
llm = ChatOpenAI(model="gpt-4o")
```

```python
# input and get response from LLM
result = llm.invoke("what is GenAI")
print(result)
```

```python
### Chat Prompt Template 
from langchain_core.prompts import ChatPromptTemplate

prompt = ChatPromptTemplate.from_messages(
    [
        ("system","You are an expert AI engineer. Provide me answer based on the quesions "),
        ("user", "{input}")
    ]
)
prompt
```

```python
# chain is combine multiple thing 
chain = prompt|llm
response = chain.invoke("input":"can you tell me about langchain")
print(response)
```

```python
type(response)  # output : AImessage
```

```python
# StrOutputParser 
from langchain_core.output_parsers import StrOutputParser
output_parser = StrOutputParser()
chain = prompt|llm|output_parser
response = chain.invoke("input":"can you tell me about langchain")
print(response)
```
