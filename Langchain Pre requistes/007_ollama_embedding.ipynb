```python
from langchain_community.embeddings import OllamaEmbeddings
```

```python
embedding = OllamaEmbeddings(model = "nomic-embed-text-v2-moe")
embedding
```

```python
r1 = embedding.embed_documents([
    "Alpha is first letter of greek alphabat",
    "Beta is Second letter of greek alphabat"
])
```

```python
len(r1[1])
```

```python
embedding.embed_query("what is second letter of greek alphabet ?")
```
